```python
import os
import re
from IPython.display import clear_output

# Setting up the contacts file
os.chdir('/content/drive/MyDrive/Colab Notebooks/file handling/contacts')
f = open('contacts.txt', mode='w')
f.write('vivek, vivek@gmail.com, 9921668254\nankur, ankur@gmail.com, 8806807004\npranav, pranav@gmail.com, 987654321\n')
f.close()

# Validation functions
def validate_name(name):
    name_pattern = re.compile('^[a-zA-Z\s]+$')
    return bool(re.match(name_pattern, name))

def validate_email(email):
    email_pattern = re.compile(r'^\w+@\w+\.[a-z]{2,3}$')
    return bool(re.match(email_pattern, email))

def validate_number(number):
    number_pattern = re.compile('^\d{10}$')
    return bool(re.match(number_pattern, number))

# Show all contacts
def show_contacts():
    try:
        f = open('contacts.txt', 'r')
        print('Contacts:')
        print(f.read())
        f.close()
    except Exception as e:
        print(e)

# Add a new contact
def add_contact():
    try:
        name = input('Enter name: ')
        if not validate_name(name):
            print("Invalid name. Name should consist of alphabets and spaces only.")
            return

        email = input('Enter email: ')
        if not validate_email(email):
            print("Invalid email.")
            return

        number = input('Enter number: ')
        if not validate_number(number):
            print("Invalid number. It should consist of 10 digits only.")
            return

        f = open('contacts.txt', 'a')
        a = name + ', ' + email + ', ' + number + '\n'
        f.write(a)
        f.close()
        print("Contact added successfully.")

    except Exception as e:
        print(e)

# Delete a contact
def delete_contact(name):
    try:
        if validate_name(name):
            f = open('contacts.txt', 'r')
            x = f.readlines()
            for i in x:
                if name in i:
                    x.remove(i)
                    print(f"Contact '{name}' deleted successfully.")
                    break
            else:
                print('No contact found with that name.')
            f.close()

            f = open('contacts.txt', 'w')
            f.writelines(x)
            f.close()

    except Exception as e:
        print(e)

# Update a contact
def update_contact(name):
    try:
        if validate_name(name):
            f = open('contacts.txt', 'r')
            a = f.readlines()
            s = ''
            for i in range(len(a)):
                if name in a[i]:
                    while True:
                        x = input('Enter n to update name, e to update email, num to update number: ').split()
                        name, email, number = a[i].strip().split(', ')
                        if 'n' in x:
                            name = input('Enter new name: ')
                        if 'e' in x:
                            email = input('Enter new email: ')
                        if 'num' in x:
                            number = input('Enter new number: ')
                        if validate_name(name) and validate_email(email) and validate_number(number):
                            b = name + ', ' + email + ', ' + number
                            update = a[i].replace(a[i], b)
                            s += update
                            print("Contact updated successfully.")
                            break
                    break
            a[i] = s + '\n'
            f.close()

            f = open('contacts.txt', 'w')
            f.writelines(a)
            f.close()
        else:
            print("Invalid name. Cannot update contact.")

    except Exception as e:
        print(e)

# Open a contact (update or delete)
def open_contact(name):
    try:
        if validate_name(name):
            x = input('Enter u to update or d to delete: ')
            if x == 'u':
                update_contact(name)
            elif x == 'd':
                delete_contact(name)
            else:
                print("Invalid choice.")
        else:
            print("Invalid name.")
    except Exception as e:
        print(e)

# Main loop
while True:
    show_contacts()

    action = input("Press 'a' - add contact, 'o' - open contact, 'q' - quit: ")
    if action == 'a':
        add_contact()
    elif action == 'o':
        name = input("Enter name of the contact you want to open: ")
        open_contact(name)
    elif action == 'q':
        print("Exiting the program.")
        break
    else:
        print("Incorrect choice.")
```
